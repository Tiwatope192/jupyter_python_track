```python
#nested loop A nested loop means a loop inside another loop.
#Each time the outer loop runs once, the inner loop runs completely.
#it can be used to see all posiible data combinatiom
#the outer loop is slower e.g a
#medium loop is medium e.g b
#inner loop is faster and runs through the other loop e.g 
```

```python
#using nested loops for crossing data
for x in range(3): #outer loop
    for y in range(2): #inner loop
        for z in range(2):
            print(f"({x}, {y}, {z}")
```

```python
#while loop repeats a block of code over and over as long as conditions are true
#differences, for loop entails iterator going through to the last value e.g (1,2,3) means 3x iteration
#while loop is iterating through unknown conditions, we do not know when it stops
```

```python
#while loops
# Build a counter from 1 to 5
count = 1            #initialization
while count <= 5:    #condition
    print(count)
    count += 1          #update
```

```python
#while loops
# Build a counter from 1 to 5
count = 1            #initialization
while count <= 5:    #condition
    print(count)
    count += 2          #update
```

```python
#write a variable that keeps asking "Do you agree?" until the user types "yes"
answer = ""
while answer != "yes":
    answer = input("Do you agree? (yes/no) : ")
    print("Thank You")
#the loop keeps going till the response is yes, then it prints thank you, no and another response is not permitted
```

```python
#Underscore is used to replace values we donot want to print 
person = ['Maria', 29, 'Data Engineer', 'Spain']
name, _, role, _ = person
print(name)
print(role)
```

```python
#combining underscore and asterisks
person = ['Maria', 29, 'Data Engineer', 'Spain']
name, *_ = person #this throws away everything after name and prints only name
print(name)
*_, country = person #this throws away everything after country and prints only country
print(country)
```

```python
#While copy adds the value z to the intended list which is copied, without altering the original
letters = ['a', 'b', 'c']
letters_copy = letters.copy()
letters_copy.append('z')
print('original:', letters)
print('copy:', letters_copy)

```

```python
#copy with pop will remove the last list from only the original
matrix = [
    ['a', 'b'],#................#row 0
     ['c', 'd'],#..............#row 1
]   
matrix_copy = matrix.copy()
matrix.pop()
print('original:', matrix)
print('copy:', matrix_copy)
```

```python
#deep copy modifies the copied list by adding an item without reflecting in the original
#copy.deepcopy()
import copy
matrix = [
    ['a', 'b'],#................#row 0
     ['c', 'd'],#..............#row 1
]   
matrix_copy = copy.deepcopy(matrix)
matrix.pop()
matrix_copy[0].append('z')
print('original:', matrix)
print('copy:', matrix_copy)
```
