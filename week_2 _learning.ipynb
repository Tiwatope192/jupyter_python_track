```python
#nested if  statements
score = 50
submitted_project = False
if score >= 90:
    if submitted_project:
        print("A+")   #....indentation is 8 spaces 
    else:
        print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else:
    print("F")
                #no condition is met, as all score were more than 50, so ouput is F
```

```python
#logical operators and, or
score = 100
submitted_project = False
if score >= 90 and submitted_project:
    print("A+")   
elif score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else:
    print("F")
                # this uses the operator "and" to combine conditions instead of multiple ifs and else
                # two of the stated conditions must be true to print the output after or else it jumps to others
                #score is more than 90 (true) but submitted assignment is false, hence A id output instead of A+
```

```python
#Break stops the loop and ends it right away, it is used for critical security risks
names = ('peju', 'tope', '', 'sokunle')
for name in names:
    if name == '':
        print('empty cell detected')
        break        #it breaks the loop once an empty cell is encounteerd, sokunle is not processed cos of the empty string
    print(f"name : {name}")
    
```

```python
#continue skips one and go to the next instead of stopping unlike breaks,#
#its the opposite of break, used for medium risks such as bad rows and empty cells
names = ('peju', 'tope', '', 'sokunle')
for name in names:
    if name == '':
        print('empty cell detected')
        continue
    print(f"name : {name}")
```

```python
#pass is a placeholder where nothing happens till you are ready to fix it
names = ('peju', 'tope', '', 'sokunle')
for name in names:
    if name == '':
        pass  #i need to handle the empty values
    print(f"name : {name}")
```

```python
#if my team agrees to replace with unknown value for processing
#pass is a placeholder where nothing happens till you are ready to fix it
names = ('peju', 'tope', '', 'sokunle')
for name in names:
    if name == '':
        #pass  #it eliminate pass and replace
        name = name.replace("", "unknown")
    print(f"name : {name}")
```
