```python
#List Comprehension comprises 3 steps
#Data Transformation, looping, filter
#Normalise the domains into standard formats
domains = ['www.google.com',
           'openai.com',
           'localhost',
           'WWW.DATAWITHBARAA.COM']
#Data cleaning and transformation
cleaned_domain = [
    #data cleaning
    d.lower().replace("www.", "")
    #looping
    for d in domains
    #data filtering
    if '.' in d
]
print(cleaned_domain)
     
```

```python
#Add .com to local host to make it a domain
domains = ['www.google.com',
           'openai.com',
           'localhost',
           'WWW.DATAWITHBARAA.COM']

cleaned_domain = [
    d.lower().replace("www.", "")
    if '.' in d else d.lower() + ".com"
    for d in domains
]

print(cleaned_domain)
```

```python
#Tuples can be ordered, indexed and allows duplicates but are immutable i.e cannot be changed unlike list
#my_tuple = (10, 30, 20, 10)
#my_tuple[3] = 40
```

```python
#Sets are unordered collection of unique items
my_set = {10, 30, 20}
print(my_set)
#it dosent allow duplicates and indexing, it takes out the duplicates in output, good for keeping clean data
my_set = {10, 30, 20, 20, 30}
print(my_set)
#it is mutable and allows changes
my_set = {10, 30, 20}
my_set.remove(20)
print(my_set)
```

```python
#set methods, adding to a set
a = {10, 20, 30, 40}
a.add(50)
print(a)
```

```python
#update() merges groups of values (iterables) into the set
a = {10, 20, 30, 40}
a.update({1,2})
print(a)
#| functions also does this 
a = {10, 20, 30, 40}
a|= ({1,2})
print(a)
```

```python
#discard() removes items if it exists and do not break the code
a = {10, 20, 30, 40}
a.discard(20)
print(a)
```

```python
#set maths method
#union() combines all unique items from both sets
a = {10, 20, 30, 40}
b = {30, 40, 50, 60}
print(a.union(b)) #or
print(a | b)
```

```python
#Intersection() returns only the shared items
print(a.intersection(b)) # or
print(a & b)

```

```python
#difference() shows items in the first set different from the second set
a = {10, 20, 30, 40}
b = {30, 40, 50, 60}
print(a.difference(b)) #or
print(a - b)
#or the reverse showing items in b different from a
print(b - a)
```

```python
#symmetric_difference() shows only non shared items
a = {10, 20, 30, 40}
b = {30, 40, 50, 60}
print(a.symmetric_difference(b)) #or
print(a ^ b)
```

```python
#set relationship method
#issubset() returns all items in this set, if it exists in the other
a = {10, 20, 30, 40}
b = {30, 40, 50, 60}
print(a.issubset(b))
a = {30, 40}
b = {30, 40, 50, 60}
print(a.issubset(b)) #everything in A can be found in b
```

```python
#issuperset() returns true when it includes all items of the other set
a = {30, 40}
b = {30, 40, 50, 60}
print(b.issuperset(a)) #b contains all the items of superset a
```

```python
#isdisjoint() returns true if both sets share no items (no overlapping)
a = {70, 80}
b = {30, 40, 50, 60}
print(a.isdisjoint(b))
```

```python
#python dictionary are ordered
my_dict = {
    'a' : 10,
    'b' : 20,
    'c' : 30
    
}
print(my_dict)
#keys must not have duplicates
my_dict = {
    'a' : 10,
    'b' : 20,
    'c' : 30,
    'a' : 40
    
}
#values can have duplicates
my_dict = {
    'a' : 10,
    'b' : 20,
    'c' : 20,
    'd' : 40
    
}
print(my_dict)
#keys are used to access values not indexes, so dicts are keyed not indexed
my_dict = {
    'a' : 10,
    'b' : 20,
    'c' : 20,
    'd' : 40
    
}
print(my_dict['b'])

#it is mutable
my_dict = {
    'a' : 10,
    'b' : 20,
    'c' : 20,
    'd' : 40
    
}
my_dict['c'] = 80
print(my_dict)
```

```python
#dict methods
user = {"id": 1, "age": 2, "city": "berlin"}
#how to access dict
print(user["city"])
#get() returns the values and gives None if missing instead of crashing the code
print(user.get("name")) #or assign an output instead of None
print(user.get("name", "unknown"))
```

```python
#In operator tests if the key is inside the dictionary
user = {"id": 1, "age": 2, "city": "berlin"}
print("age" in user)
print("name" in user)
```

```python
#View objects gives a live view of dict keys, values or key value pairs
print(user.keys())
print(user.values())
#items prints both key and values, ideal for looping,transformation, comparism and building new dicts
print(user.items())
#looping
for key, value in user.items():
    print(key, value)
```

```python
#Add, Remove, update
user = {"id": 1, "age": 2, "city": "berlin"}
user["name"] = "John" #adds
print(user)
user["age"] = 35 #updates
print(user) #or
#update() adds new keys and updates existing ones using another dict
user.update({"age": 40, "city": "Paris"})
print(user)
#pop() removes a key from the dictionary and returns its value
age = user.pop("age")
print(user)
print("Removed Item:", age)
#In order not to crash the code, assign an output if key is missing)
age = user.pop("gender", "Not Found")
print(user)
print("Removed Item:", age)
#popitem() returns and delete the most recent key value pair from the dict
user.popitem()
print(user)
```
