```python
#Given two lists of students who attended different classes
class_a = ["John", "Mary", "Peter", "John", "Sarah"]
class_b = ["Peter", "Daniel", "Mary", "Mary", "Paul"]

#Tasks:
#Convert both to sets
#Remove duplicates automatically
#Find students that attended both classes
#Find students that attended only classes A

#1 convert both to sets #2 Print all results and Remove duplicates automatically
set_a = set(class_a)
set_b = set(class_b)
students = set_a | set_b
print(students)

#3 Find students that attended both classes
print(set_a.intersection(set_b))

#4 Find students that attended only classes a
print(set_a.difference(set_b))
```

```python
#Print numbers from 1 to 20 using a for loop.
for numbers in range (1, 21):
    print(numbers)

#Print only even numbers from 1 to 20.
for numbers in range(1, 21):
    if numbers % 2 == 0:
        print(numbers)

#Print the square of each number from 1 to 10.
for numbers in range(1, 11):
        print(numbers ** 2)

#Given a list [4, 7, 1, 9, 3], print each element.
numbers = [4, 7, 1, 9, 3]
for number in numbers:
    print(number)
#Count how many numbers in a list are greater than 5.
numbers = [4, 7, 1, 9, 3]
count = 0
for number in numbers:
    if number > 5:
        count += 1
print(count)
#Find the sum of all elements in a list without using sum().
numbers = [1, 2, 3, 4, 5]
Total = 0
for number in numbers:
    Total += number 
print("Sum:", Total)
#Find the maximum value in a list without using max().
numbers = [1, 2, 3, 4, 5]
largest = numbers[0]
for number in numbers:
    if number > largest:
        largest = number
print("Maximum:", largest)
```

```python
#Write a Python function that accepts two integer numbers. 
#If the product of the two numbers is less than or equal to 1000, return their product; otherwise, return their sum.
def calculate_result(num1, num2):
    product = num1 * num2
    
    if product <= 1000:
        return product
    else:
        return num1 + num2




  
```

```python
#Iterate through the first 10 numbers (0–9). In each iteration, print the current number, the previous number, and their sum.

previous_num = 0

for current_num in range(0, 10):
    total = current_num + previous_num
    print("Current:", current_num,
          "Previous:", previous_num,
          "Sum:", total)
    
    previous_num = current_num
```

```python
#Write a program that takes a string and reverses it (e.g., “Python” becomes “nohtyP”).

# Write a program that takes a string and reverses it (e.g., "Python" becomes "nohtyP").
text = "python"
reversed_text = text[::-1]  # Added 'text' before the slice notation
print(f"Original: {text}")
print(f"Reversed: {reversed_text}")
```

```python
 #Write a program to count the total number of vowels (a, e, i, o, u) present in a given sentence "python is amazing"
sentence = "python is amazing"
vowels = "aeiou"
count = 0
for char in sentence.lower(): #Go through the sentence one character at a time.”
    if char in vowels:        #Check if the current character is inside the string "aeiou".”
        count += 1 #If the condition is true, increase the vowel count
print("Total vowels:", count)


```

```python
 #Given a list of integers, find and print both the largest and the smallest numbers.
nums = [45, 2, 89, 12, 7]
largest = max(nums)
smallest = min(nums)    
print("largest:", largest)
print("smallest:", smallest)
```

```python
 #Write a function to return True if the first and last number of a given list is the same. If the numbers are different, return False
numbers_x = [10, 20, 30, 40, 10]
numbers_y = [75, 65, 35, 75, 30]
def check_first_last(numbers):
    if numbers[0] == numbers[-1]:
        return True
    else:
        return False
print(check_first_last(numbers_x))
print(check_first_last(numbers_y))
```
