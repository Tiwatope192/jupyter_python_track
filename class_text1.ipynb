```python
#renaming phone number
df.rename(columns={"Phone 1": "primary_number", "Phone 2": "secondary_number"}, inplace=True)
```

```python
#using the re package similar to regex in SQL to match validate and change strings
import re
```

```python
#creating a function to standardising the phone number
def format_number_safe(phone_number):
    if pd.isna(phone_number):
        return None
    cleaned = re.sub(r"\D", "", str(phone_number))
    if len(cleaned) == 11 and cleaned.startswith("1"):
        cleaned = cleaned[1:]
    if len(cleaned) == 10:
        return f"+1 {cleaned[:3]} {cleaned[3:6]} {cleaned[6:]}"
    return None

```

```python
df["primary_number"] = df["primary_number"].apply(format_number_safe)
df["secondary_number"] = df["secondary_number"].apply(format_number_safe)
```

```python
#store into a new data frame
df_cleaned = df.copy()  # or df_cleaned = df[["name", "primary_number", "secondary_number"]]

print(df_cleaned)
```
